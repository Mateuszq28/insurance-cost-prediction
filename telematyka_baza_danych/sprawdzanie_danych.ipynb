```python
import pandas as pd

df = pd.read_csv('med-expense.csv')
df['region']
```

```python
for index, row in df.iterrows():
    df.iloc[index, 1] = int(row.iloc[1] == "male")
    df.iloc[index, 4] = int(row.iloc[4] == "yes")

print(df)
```

```python
from sklearn.preprocessing import OneHotEncoder

model_enc = OneHotEncoder()
model_enc = model_enc.fit(df.iloc[:,5:6])
print(model_enc.categories_)
model_enc = model_enc.transform(df.iloc[:,5:6])
model_enc_arr = model_enc.toarray()
print(model_enc_arr[-6:])
```

```python
df.drop(columns='region', inplace=True)
```

```python

df = df.join(pd.DataFrame(
    {
        'is_southwest': model_enc_arr[:,3],
        'is_southeast': model_enc_arr[:,2],
        'is_northwest': model_enc_arr[:,1],
        'is_northeast': model_enc_arr[:,0]
    }, index=df.index
))
```

```python
df
```

```python
df.rename(columns = {'sex':'is_man', 'smoker':'is_smoking'}, inplace = True)
cols = df.columns.tolist()
cols = cols[:5] + cols[6:] + cols[5:6]
df = df[cols]
df
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]

x
```

```python
y
```

```python
x = x.to_numpy()
x
```

```python
y = y.to_numpy()
y
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x, y)
```

```python
# model.predict(x.iloc[0:1,:])
model.predict(x[0:1,:])
```

```python
y_pred = model.predict(x)
```

```python
from sklearn.metrics import mean_squared_error as MSE

MSE(y, y_pred)
```
